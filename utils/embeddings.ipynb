```python
import pymongo
from dotenv import load_dotenv
import os

from langchain_openai import OpenAIEmbeddings
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.chat_models import ChatAnthropic
from langchain_openai import ChatOpenAI, OpenAI
from langchain_mistralai.chat_models import ChatMistralAI


from langchain.prompts import PromptTemplate
# from langchain.output_parsers import PydanticOutputParser
from langchain_core.output_parsers import JsonOutputParser, StrOutputParser, PydanticOutputParser
import os
load_dotenv()

os.environ['LANGCHAIN_TRACING_V2'] = 'true'
```

```python
client = pymongo.MongoClient(os.getenv('mongodb_uri'))
DB_NAME = "memories"
COLLECTION_NAME = "memoriescollections"
ATLAS_VECTOR_SEARCH_INDEX_NAME = "vector_index"
MONGODB_COLLECTION = client[DB_NAME][COLLECTION_NAME]
```

```python
for item in MONGODB_COLLECTION.find():
    print(item)
```

```python
from langchain.embeddings import OpenAIEmbeddings
openai_embeddings = OpenAIEmbeddings()

for item in MONGODB_COLLECTION.find():
    raw_text_response = item['rawTextResponse']
    print(raw_text_response)
    text_response_embedding = openai_embeddings.embed_query(raw_text_response)
    MONGODB_COLLECTION.update_one(
        {'_id': item['_id']},
        {'$set': {'textResponseEmbedding': text_response_embedding}}
    )
```
